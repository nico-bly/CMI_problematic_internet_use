```python
from concurrent.futures import ThreadPoolExecutor

def process_file(filename, dirname):
    df = pd.read_parquet(os.path.join(dirname, filename, 'part-0.parquet'))
    df.drop('step', axis=1, inplace=True)
    return df.describe().values.reshape(-1), filename.split('=')[1]

def load_time_series(dirname) -> pd.DataFrame:
    ids = os.listdir(dirname)
    
    with ThreadPoolExecutor() as executor:
        results = list(tqdm(executor.map(lambda fname: process_file(fname, dirname), ids), total=len(ids)))
    
    stats, indexes = zip(*results)
    
    df = pd.DataFrame(stats, columns=[f"stat_{i}" for i in range(len(stats[0]))])
    df['id'] = indexes
    
    return df

#train_ts = load_time_series(train_path_ts)
#test_ts = load_time_series(test_path_ts)
```

```python
import pandas as pd
import os
from pathlib import Path

def load_time_series(data_paths, id_):
    try:
        # Construire le chemin correct vers le fichier parquet
        file_path = os.path.join(data_paths, f"id={id_}", "part-0.parquet")
        #print(file_path)
        
        # Vérifier si le fichier existe
        if not os.path.exists(file_path):
            #print(f"Fichier non trouvé pour l'ID: {id_}")
            return None
            
        # Lire le fichier parquet
        return pd.read_parquet(file_path)
    except Exception as e:
        #print(f"Erreur lors du chargement de l'ID {id_}: {str(e)}")
        return None
```

```python

# Création d'un dictionnaire pour stocker les séries temporelles
time_series_dict = {}
path=data_paths / 'series_train.parquet'
# Chargement des données pour chaque ID
for idx, row in df.iterrows():
    ts = load_time_series(path, row['id'])
    if ts is not None:
        time_series_dict[row['id']] = {
            'series': ts,
            'target': row['sii']
        }
```

```python
import torch
import torch.nn as nn

class TimeSeriesClassifier(nn.Module):
    def __init__(self, input_dim=12, hidden_dim=64, num_layers=2, num_classes=6):
        super().__init__()
        self.lstm = nn.LSTM(
            input_size=input_dim,
            hidden_size=hidden_dim,
            num_layers=num_layers,
            batch_first=True,
            dropout=0.2
        )
        
        self.classifier = nn.Sequential(
            nn.Linear(hidden_dim, 32),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(32, num_classes)
        )
        
    def forward(self, x):
        lstm_out, _ = self.lstm(x)
        # Prendre uniquement la dernière sortie
        last_hidden = lstm_out[:, -1, :]
        return self.classifier(last_hidden)

# Préparation des données
def prepare_data(time_series_dict, sequence_length=100):
    X, y = [], []
    
    for id_, data in tqdm(time_series_dict.items(), desc="Préparation des données"):
        series = data['series']
        features = series[['X', 'Y', 'Z', 'enmo', 'anglez', 'light']]
        
        # Normalisation
        features = (features - features.mean()) / features.std()
        
        # Création des séquences
        for i in range(0, len(features) - sequence_length, sequence_length):
            X.append(features[i:i+sequence_length].values)
            y.append(data['target'])
        
    print('yolo')   
    return torch.FloatTensor(X), torch.LongTensor(y)

def train_model(model, train_loader, num_epochs=10, learning_rate=0.001):
    criterion = nn.CrossEntropyLoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)
    
    for epoch in tqdm(range(num_epochs), desc="Epochs"):
        for batch_X, batch_y in tqdm(train_loader, desc=f"Batch Epoch {epoch+1}", leave=False):
            optimizer.zero_grad()
            outputs = model(batch_X)
            loss = criterion(outputs, batch_y)
            loss.backward()
            optimizer.step()
            
        # Afficher la loss à la fin de chaque epoch
        print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}')
```

```python
# Création du modèle
model = TimeSeriesClassifier(
    input_dim=6,  # X, Y, Z, enmo, anglez, light
    hidden_dim=64,
    num_layers=2,
    num_classes=6  # Classes de 0 à 5 pour les niveaux d'utilisation d'internet
)

# Préparation des données
X, y = prepare_data(time_series_dict, sequence_length=100)

# Création du DataLoader
from torch.utils.data import TensorDataset, DataLoader
dataset = TensorDataset(X, y)
train_loader = DataLoader(dataset, batch_size=32, shuffle=True)

# Entraînement
train_model(
    model=model,
    train_loader=train_loader,
    num_epochs=10,
    learning_rate=0.001
)
```
